How many total rows are there?

To find false alarms between certain hours use the WHERE command and join several clauses carefully with 'AND' and 'OR'


How many False Alarms happened between 10PM and 5 AM?

```python
%%bigquery --project data-visualization-class-sql
SELECT COUNT(*) as number_of_late_night_false_alarms
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
WHERE incident_group = "False Alarm" AND (hour_of_call >= 22 OR hour_of_call<=4)
```

What is the average number of num_pumps_attending when there is a Fire?

If we want to see about when there is a fire, we will need to either restrict our data to then or possible GROUP BY . We will use GROUP BY since we already did one with the WHERE.

```python
%%bigquery --project data-visualization-class-sql
SELECT INCIDENT_GROUP, AVG(num_pumps_attending) as num_pumps_avg
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
GROUP BY incident_group
```

What is the largest number of num_pumps_attending to a Road Vehicle property_category

```python
%%bigquery --project data-visualization-class-sql
SELECT MAX(num_pumps_attending) as num_pumps_max, property_category 
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
GROUP BY property_category
```
