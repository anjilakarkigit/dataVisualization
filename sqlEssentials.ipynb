I am going to get the average time for a trip based on starting point using a GROUP BY command.

```python
%%bigquery --project data-visualization-class-sql
SELECT  start_station_name, AVG(duration_minutes) as avg_duration
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
```

```python
%%bigquery --project data-visualization-class-sql
SELECT start_station_name, COUNT( start_station_name) as start_station_count
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY start_station_count DESC

```
