```python
def JuventusGoals(row):
  return row.Goals

def JuventusAssists(row):
  return row.Assists
```

```python
df2 = df[df['club'].str.contains('Juventus')]

df['JU_goals'] = df2.apply(lambda row: JuventusGoals(row), axis=1)

df['JU_assists'] = df2.apply(lambda row: JuventusAssists(row), axis=1)
```

```python
df['JU_goals'] = df['JU_goals'].fillna(0).astype(int)
df['JU_goals'] = df['JU_goals'].astype('int')
df['JU_assists'] = df['JU_assists'].fillna(0).astype(int)
df['JU_assists'] = df['JU_assists'].astype('int')
df[df['club'].str.contains('Juventus')]
```

```python
sns.relplot(data = df, x = 'Goals', y = 'Assists', color = 'green', style = 'JU_goals', size = 'JU_assists')

plt.title('A scatterplot of goals vs assist')
```

#H

```python
df3 = df[df['nationality'].str.contains('France')]
ax = df3[['Goals', 'Assists']].plot(kind = 'bar', stacked = True,figsize=(15, 10) , title = 'Goals and Assist of French National Team')
ax.set_xlabel('Goals')
ax.set_ylabel('Assists')
```

The barplot of Assists and Goals for the France National Team is shown above. The major purpose of stacked bar charts is to convey the big picture of data while still allowing you to view the minute details that influence the big image. Individual goals and assists, as well as the total number of both, may be found here. Goals and assists are relatable because, as shown in the graph above, players score the same amount of goals as they assist. With the most goals and assists, it appears that Florian Thauvin contributed the most to the team's scoring.
