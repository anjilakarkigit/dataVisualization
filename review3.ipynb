```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns

df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/worldCupPlayer.csv')
df.head()
```

```python
ax = sns.scatterplot(data = df, x = 'Rating', y = 'age', hue = 'MotM')
ax.set(title = "Scatter")

plt.show()
```

```python
ax = sns.heatmap(df.corr(),vmin = -1)
ax.set(title = 'Heatmap of Worldcup Players')

plt.show()
```
