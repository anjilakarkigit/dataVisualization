Now that I am authenticated, I can start to play around in the dataset. I am going to look at dataset from 'austin_bikeshare.bikeshare_trips' and try to find the number of entries, the longest trip from 'durration_mintues' and average time for the trip from the dataset. 

```python
%%bigquery --project data-visualization-class-sql
SELECT 
  COUNT(*) as total_rows
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
```

```python
%%bigquery --project data-visualization-class-sql 
SELECT 
  MAX(duration_minutes) as max_minute
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
```

```python
%%bigquery --project data-visualization-class-sql 
SELECT 
  AVG(duration_minutes) as avg_minute
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
```

```python
%%bigquery --project data-visualization-class-sql
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 5

```
