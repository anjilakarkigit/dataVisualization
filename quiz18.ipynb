```python
import pandas as pa

df = pa.read_csv("https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv")
df.head()
```

First, I will be converting the room type to its appropriate numerical data.I will be using get dummies function to convert it to the numerical value.

```python
pa.get_dummies(df.room_type).head()
```

Here, we can see the room types indicated as 1 and 0. 1 means true and 0 means false.

```python
df1 = pa.concat([df,pa.get_dummies(df.room_type)],axis = 1)

df1 = df1.drop('room_type', axis = 1)

df1.head()
```

The room type has been transformed to numerical data in the table. The dummy function was easy to use for room type because it only has three types, which implies there isn't a lot of data.

I think to convert the neighbour group I would have to clean up some data.

```python
df = pa.read_csv("https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv")
df.head()
```

```python
edlevels = sorted(df.neighbourhood_group.unique())

edlevels
```

```python
eddict = {}

for i in range(5):
  eddict[edlevels[i]] = i

eddict
```

```python
df.groupby('neighbourhood_group').host_id.agg('count')
```

```python
df.neighbourhood_group.map(eddict)
```

```python
df.neighbourhood_group = df.neighbourhood_group.map(eddict)

df.head()
```

As you can see, I transformed the numerical data separately because I didn't know how to convert it in the same table and it was difficult. We used a for loop to assign a numerical value to each neighborhood group, then used the groupby function to group them and convert them.
