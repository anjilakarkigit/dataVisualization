What is the subscrriber_type for most popular start station and is the most popular subscriber type also popular in most popular start station?


```python
%%bigquery --project data-visualization-class-sql
SELECT COUNT(start_station_name) as count, start_station_name as name, subscriber_type as subs
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY name, subs
ORDER BY count DESC
```

```python
%%bigquery --project data-visualization-class-sql
SELECT COUNT(subscriber_type) as count,subscriber_type as subs, start_station_name as name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY subs,name
ORDER BY count DESC
```

Lets find the most popular start station and the subscriber type for it.

```python
%%bigquery --project data-visualization-class-sql
SELECT start_station_name,COUNT(start_station_name) as start_station_count, subscriber_type as subscriber
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name, subscriber
ORDER BY start_station_count DESC


```

```python
From the above table, it is seen that the most popular start station is 21st & Speedway @PCL. And the subscriber_type for it is UT Student memberrship. 
now lets find out the most popular subscriber type and its corresponding start_station_name
```

```python
%%bigquery --project data-visualization-class-sql
SELECT start_station_name, COUNT(subscriber_type) as subscribercount, subscriber_type as subscriber
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name, subscriber
ORDER BY subscribercount DESC
```

```python
%%bigquery --project data-visualization-class-sql
SELECT COUNT(subscriber_type) as subscriber_count, subscriber_type as subscriber
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY subscriber
ORDER BY subscriber_count DESC
```

```python


%%bigquery --project data-visualization-class-sql
WITH statsTable as(
SELECT COUNT(subscriber_type) as subscriber_count, subscriber_type as subscriber
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
)

SELECT t.start_station_name, COUNT(t.start_station_name) as station_count
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
      ON t.start_station_name = statsTable.start_station_name
GROUP BY start_station_name
ORDER BY station_count DESC

```

```python
%%bigquery --project data-visualization-class-sql

SELECT   trips.subscriber_type as subscriber, station.start_station_name, COUNT(station.start_station_name) as station_count
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`as trips LEFT OUTER JOIN `bigquery-public-data.austin_bikeshare.bikeshare_trips` as station
      ON trips.start_station_name= station.start_station_name
      GROUP BY  start_station_name
      ORDER BY station_count
LIMIT 10
```
