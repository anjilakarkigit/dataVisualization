```python

%%bigquery --project data-visualization-class-sql
SELECT name, property_type
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
LIMIT 10
```

```python
%%bigquery --project data-visualization-class-sql

SELECT station.name, station.property_type, trips.duration_minutes
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`as trips LEFT OUTER JOIN `bigquery-public-data.austin_bikeshare.bikeshare_stations` as station
      ON trips.start_station_name= station.name

LIMIT 4
```

```python
%%bigquery --project data-visualization-class-sql

SELECT  station.property_type, AVG(trips.duration_minutes ) as avergaetime
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`as trips LEFT OUTER JOIN `bigquery-public-data.austin_bikeshare.bikeshare_stations` as station
      ON trips.start_station_name= station.name
      GROUP BY  property_type
```
